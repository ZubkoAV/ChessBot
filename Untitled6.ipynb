```python
import chess
import chess.svg
import IPython
from IPython.display import display, SVG, clear_output, HTML
import ipywidgets as widgets
from ipywidgets import interact, IntSlider
import random
import time
import copy
import chess.polyglot
from concurrent.futures import ProcessPoolExecutor
import multiprocessing
import json
import os
import threading
from google.colab import output
```

```python
class GeneticBot:
    def __init__(self, name="GenBot", genome=None, time_limit=0.5):
        self.name = name
        self.time_limit = time_limit
        self.eval_cache = {}
        self.killer_moves = {}
        self.nodes_searched = 0

        if genome is None:
            self.genome = {
                'pawn': random.uniform(80, 120),
                'knight': random.uniform(280, 360),
                'bishop': random.uniform(290, 370),
                'rook': random.uniform(450, 550),
                'queen': random.uniform(800, 1000),
                'mobility': random.uniform(0.5, 3.0),
                'knight_center': random.uniform(10, 40),
                'pawn_center': random.uniform(10, 30),
                'bishop_pair': random.uniform(30, 80),
                'passed_pawn': random.uniform(20, 60),
                'king_safety': random.uniform(10, 50)
            }
        else:
            self.genome = genome.copy()

    def evaluate_board(self, board):
        hash_key = chess.polyglot.zobrist_hash(board)
        if hash_key in self.eval_cache:
            return self.eval_cache[hash_key]

        if board.is_checkmate():
            return -99999

        if board.is_stalemate() or board.is_insufficient_material():
            return 0

        score = 0
        center_squares = {chess.D4, chess.E4, chess.D5, chess.E5}

        piece_values = {
            chess.PAWN: self.genome['pawn'],
            chess.KNIGHT: self.genome['knight'],
            chess.BISHOP: self.genome['bishop'],
            chess.ROOK: self.genome['rook'],
            chess.QUEEN: self.genome['queen'],
            chess.KING: 20000
        }

        white_bishops = black_bishops = 0

        for square in chess.SQUARES:
            piece = board.piece_at(square)
            if piece:
                value = piece_values[piece.piece_type]

                # Позиційні бонуси
                if piece.piece_type == chess.KNIGHT and square in center_squares:
                    value += self.genome['knight_center']
                elif piece.piece_type == chess.PAWN and square in center_squares:
                    value += self.genome['pawn_center']
                elif piece.piece_type == chess.BISHOP:
                    if piece.color == chess.WHITE: white_bishops += 1
                    else: black_bishops += 1

                if piece.color == board.turn: score += value
                else: score -= value

        if white_bishops >= 2: score += self.genome['bishop_pair'] if board.turn == chess.WHITE else -self.genome['bishop_pair']
        if black_bishops >= 2: score += self.genome['bishop_pair'] if board.turn == chess.BLACK else -self.genome['bishop_pair']

        score += board.legal_moves.count() * self.genome['mobility']

        self.eval_cache[hash_key] = score
        return score

    def order_moves(self, board, depth, only_captures=False):
        moves = list(board.generate_legal_captures()) if only_captures else list(board.legal_moves)

        killers = self.killer_moves.get(depth, [])

        def move_score(move):
            score = 0
            if board.is_capture(move):
                victim = board.piece_at(move.to_square)
                attacker = board.piece_at(move.from_square)
                if victim and attacker:
                    score += 100 * victim.piece_type - attacker.piece_type
            elif move in killers:
                score += 50

            if move.promotion: score += 900
            return score

        return sorted(moves, key=move_score, reverse=True)

    def quiescence_search(self, board, alpha, beta):
        self.nodes_searched += 1
        stand_pat = self.evaluate_board(board)

        if stand_pat >= beta: return beta
        if alpha < stand_pat: alpha = stand_pat

        for move in self.order_moves(board, depth=0, only_captures=True):
            board.push(move)
            score = -self.quiescence_search(board, -beta, -alpha)
            board.pop()

            if score >= beta: return beta
            if score > alpha: alpha = score

        return alpha

    def negamax(self, board, depth, ply_from_root, alpha, beta, end_time):
        if time.time() > end_time:
            raise TimeoutError()

        self.nodes_searched += 1

        if depth == 0 or board.is_game_over():
            return self.quiescence_search(board, alpha, beta)

        ordered_moves = self.order_moves(board, depth)
        if not ordered_moves:
            return self.evaluate_board(board)

        max_val = float('-inf')

        for move in ordered_moves:
            board.push(move)
            try:
                score = -self.negamax(board, depth - 1, ply_from_root + 1, -beta, -alpha, end_time)
            finally:
                board.pop()

            if score > max_val: max_val = score
            if score > alpha: alpha = score

            if alpha >= beta:
                if not board.is_capture(move):
                    if depth not in self.killer_moves: self.killer_moves[depth] = []
                    if move not in self.killer_moves[depth]:
                        self.killer_moves[depth].insert(0, move)
                        self.killer_moves[depth] = self.killer_moves[depth][:2]
                break

        return max_val

    def choose_move(self, board):
        self.eval_cache.clear()
        self.killer_moves.clear()
        self.nodes_searched = 0

        start_time = time.time()
        end_time = start_time + self.time_limit

        best_move = None
        best_move_overall = None

        legal_moves = self.order_moves(board, depth=0)
        if not legal_moves: return None

        if len(legal_moves) == 1: return legal_moves[0]

        try:
            for current_depth in range(1, 10):
                best_value = float('-inf')
                alpha = float('-inf')
                beta = float('inf')

                if best_move_overall:
                    if best_move_overall in legal_moves:
                        legal_moves.remove(best_move_overall)
                        legal_moves.insert(0, best_move_overall)

                for move in legal_moves:
                    board.push(move)
                    try:
                        board_value = -self.negamax(board, current_depth - 1, 1, -beta, -alpha, end_time)
                    finally:
                        board.pop()

                    if board_value > best_value:
                        best_value = board_value
                        best_move = move

                    alpha = max(alpha, best_value)

                best_move_overall = best_move

        except TimeoutError:
            pass

        return best_move_overall if best_move_overall else random.choice(legal_moves)

    @staticmethod
    def crossover(parent1, parent2):
        child_genome = {}
        for key in parent1.genome.keys():
            child_genome[key] = parent1.genome[key] if random.random() < 0.5 else parent2.genome[key]
        return child_genome

    @staticmethod
    def mutate(genome, mutation_rate=0.15, mutation_strength=0.15):
        mutated = genome.copy()
        for key in mutated.keys():
            if random.random() < mutation_rate:
                mutated[key] *= random.uniform(1 - mutation_strength, 1 + mutation_strength)
                mutated[key] = max(0, mutated[key])
        return mutated

def play_match(args):
    bot1_genome, bot2_genome, game_num = args
    bot1 = GeneticBot("White", genome=bot1_genome, time_limit=0.1)
    bot2 = GeneticBot("Black", genome=bot2_genome, time_limit=0.1)

    board = chess.Board()
    move_count = 0

    while not board.is_game_over(claim_draw=True) and move_count < 80:
        bot = bot1 if board.turn == chess.WHITE else bot2
        move = bot.choose_move(board)
        if move:
            board.push(move)
            move_count += 1
        else:
            break

    if board.is_checkmate():
        return 1.0 if not board.turn else 0.0
    return 0.5

def tournament_selection(population, fitness_scores, tournament_size=3):
    selected_indices = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected_indices, key=lambda idx: fitness_scores[idx])
    return population[best_idx]

def evolve_population(population_size=10, generations=5, games_per_matchup=2):
    print(f"Початок паралельної еволюції: {population_size} ботів, {generations} поколінь")
    print(f"Використовується ядер CPU: {multiprocessing.cpu_count()}\n")

    population = [GeneticBot(f"Bot_{i}") for i in range(population_size)]

    for gen in range(generations):
        print(f"{'='*40}\nПокоління {gen + 1}/{generations}\n{'='*40}")
        fitness = [0.0] * population_size

        matchups = []
        matchup_indices = []

        for i in range(population_size):
            for j in range(i + 1, population_size):
                for game_num in range(games_per_matchup):
                    if game_num % 2 == 0:
                        matchups.append((population[i].genome, population[j].genome, game_num))
                        matchup_indices.append((i, j)) # i грає білими
                    else:
                        matchups.append((population[j].genome, population[i].genome, game_num))
                        matchup_indices.append((j, i)) # j грає білими

        with ProcessPoolExecutor() as executor:
            results = list(executor.map(play_match, matchups))

        for idx, result in enumerate(results):
            white_idx, black_idx = matchup_indices[idx]
            if result == 1.0: fitness[white_idx] += 1.0
            elif result == 0.0: fitness[black_idx] += 1.0
            else:
                fitness[white_idx] += 0.5
                fitness[black_idx] += 0.5

        ranked = sorted(enumerate(fitness), key=lambda x: x[1], reverse=True)
        print("Топ-3 боти:")
        for idx, (bot_idx, score) in enumerate(ranked[:3]):
            print(f"  {idx+1}. {population[bot_idx].name}: {score:.1f} балів")

        if gen < generations - 1:
            new_population = [population[ranked[0][0]], population[ranked[1][0]]] # Елітизм
            while len(new_population) < population_size:
                parent1 = tournament_selection(population, fitness)
                parent2 = tournament_selection(population, fitness)
                child_genome = GeneticBot.mutate(GeneticBot.crossover(parent1, parent2))
                new_population.append(GeneticBot(f"Bot_Gen{gen+2}_{len(new_population)}", child_genome))
            population = new_population

    best_idx = max(enumerate(fitness), key=lambda x: x[1])[0]
    best_bot = population[best_idx]
    print("\nЕволюція завершена! Найкращий геном:")
    for k, v in best_bot.genome.items(): print(f"  {k}: {v:.1f}")
    return best_bot

```

```python
# ==========================================
# 1. ФУНКЦІЇ ЗБЕРЕЖЕННЯ ТА ЗАВАНТАЖЕННЯ
# ==========================================
def save_genome(genome, filename="bot_weights.json"):
    """Зберігає геном бота у файл."""
    with open(filename, 'w', encoding='utf-8') as f:
        json.dump(genome, f, indent=4)
    print(f"✅ Геном успішно збережено у файл '{filename}'")

def load_genome(filename="bot_weights.json"):
    """Завантажує геном з файлу, якщо він існує."""
    if os.path.exists(filename):
        with open(filename, 'r', encoding='utf-8') as f:
            return json.load(f)
    return None


# ==========================================
# 2. ІНІЦІАЛІЗАЦІЯ БОТА (АБО НАВЧАННЯ)
# ==========================================
file_name = "bot_weights.json"
print("Ініціалізація шахового ядра...")

saved_genome = load_genome(file_name)

if saved_genome:
    print("🤖 Файл з вагами знайдено! Бот 'Titan' завантажений і готовий до гри.")
    current_bot = GeneticBot(name="Titan", genome=saved_genome, time_limit=1.5)
else:
    print("⚠️ Файл з вагами не знайдено! Спершу потрібно натренувати бота.")
    print("⏳ Запускаємо процес еволюції...")

    champion_bot = evolve_population(population_size=10, generations=10, games_per_matchup=2)

    save_genome(champion_bot.genome, file_name)

    current_bot = GeneticBot(name="Titan", genome=champion_bot.genome, time_limit=1.5)


# ==========================================
# 3. ГЛОБАЛЬНІ ЗМІННІ ТА ФУНКЦІЇ СТАНУ
# ==========================================
play_board = chess.Board()
player_color = chess.WHITE

def get_status_message(board):
    if board.is_checkmate():
        winner = "Білі" if board.outcome().winner == chess.WHITE else "Чорні"
        return f"🏆 МАТ! Перемогли {winner}."
    if board.is_stalemate():
        return "🤝 НІЧИЯ! Пат."
    if board.is_insufficient_material() or board.can_claim_draw():
        return "🤝 НІЧИЯ!"
    if board.is_game_over():
        return f"🏁 Гра завершена."
    return "✅ Твій хід..."


# ==========================================
# 4. BACKEND: ОБРОБНИК ХОДІВ
# ==========================================
def handle_move(move_str):
    global play_board, player_color

    try:
        move = chess.Move.from_uci(move_str)
    except:
        return IPython.display.JSON({'fen': play_board.fen(), 'status': '❌ Помилка формату'})

    if move in play_board.legal_moves:
        play_board.push(move)

        if play_board.is_game_over(claim_draw=True):
            return IPython.display.JSON({'fen': play_board.fen(), 'status': get_status_message(play_board)})

        bot_move = current_bot.choose_move(play_board)
        if bot_move:
            play_board.push(bot_move)

        return IPython.display.JSON({'fen': play_board.fen(), 'status': get_status_message(play_board)})
    else:
        return IPython.display.JSON({'fen': play_board.fen(), 'status': '❌ Нелегальний хід!'})

output.register_callback('notebook.handle_move', handle_move)


# ==========================================
# 5. FRONTEND: ВІДМАЛЬОВКА ДОШКИ (HTML/JS)
# ==========================================
def start_new_game(b=None):
    global play_board, player_color
    play_board = chess.Board()

    choice = color_dropdown.value
    if choice == 'Випадково':
        player_color = random.choice([chess.WHITE, chess.BLACK])
    elif choice == 'Білі':
        player_color = chess.WHITE
    else:
        player_color = chess.BLACK

    board_orientation = 'white' if player_color == chess.WHITE else 'black'
    initial_status = "✅ Твій хід..."

    if player_color == chess.BLACK:
        bot_move = current_bot.choose_move(play_board)
        if bot_move:
            play_board.push(bot_move)
        initial_status = get_status_message(play_board)

    html_code = f"""
    <link rel="stylesheet" href="https://unpkg.com/@chrisoakman/chessboardjs@1.0.0/dist/chessboard-1.0.0.min.css" />
    <script src="https://code.jquery.com/jquery-3.5.1.min.js"></script>
    <script src="https://unpkg.com/@chrisoakman/chessboardjs@1.0.0/dist/chessboard-1.0.0.min.js"></script>

    <div style="display: flex; flex-direction: column; align-items: center; padding: 20px; font-family: sans-serif;">
        <div id="board_div" style="width: 450px; margin: 10px 0; box-shadow: 0px 4px 15px rgba(0,0,0,0.3); border-radius: 4px;"></div>
        <h3 id="statusText" style="margin-top: 15px; color: #1a73e8; min-height: 24px; font-weight: bold;">{initial_status}</h3>
    </div>

    <script>
    setTimeout(function() {{
        var config = {{
          draggable: true,
          position: '{play_board.fen()}',
          orientation: '{board_orientation}',
          pieceTheme: 'https://chessboardjs.com/img/chesspieces/wikipedia/{{piece}}.png',

          onDragStart: function(source, piece, position, orientation) {{
              var status = document.getElementById('statusText').innerText;
              if (status.includes('МАТ') || status.includes('НІЧИЯ') || status.includes('завершена')) return false;
              if ((orientation === 'white' && piece.search(/^b/) !== -1) ||
                  (orientation === 'black' && piece.search(/^w/) !== -1)) return false;
          }},

          onDrop: async function(source, target, piece, newPos, oldPos, orientation) {{
              if (source === target) return 'snapback';
              var move = source + target;
              if (piece === 'wP' && target[1] === '8') move += 'q';
              if (piece === 'bP' && target[1] === '1') move += 'q';

              document.getElementById('statusText').innerText = "🤖 TitanBot думає...";

              try {{
                  const result = await google.colab.kernel.invokeFunction('notebook.handle_move', [move], {{}});
                  const data = result.data['application/json'];
                  window.myBoard.position(data.fen);
                  document.getElementById('statusText').innerText = data.status;
              }} catch(e) {{
                  document.getElementById('statusText').innerText = "❌ Помилка з'єднання.";
                  return 'snapback';
              }}
          }}
        }};
        window.myBoard = Chessboard('board_div', config);
    }}, 200);
    </script>
    """
    with output_area:
        IPython.display.clear_output(wait=True)
        IPython.display.display(IPython.display.HTML(html_code))

# ==========================================
# 6. UI: ВІДЖЕТИ ЗАПУСКУ
# ==========================================
color_dropdown = widgets.Dropdown(options=['Білі', 'Чорні', 'Випадково'], value='Білі', description='Колір:')
start_btn = widgets.Button(description='Почати нову гру', button_style='success', icon='play')

start_btn.on_click(start_new_game)
output_area = widgets.Output()

display(widgets.HBox([color_dropdown, start_btn]))
display(output_area)

start_new_game(None)
```

```python
class GeneticBot:
    def __init__(self, name="GenBot", genome=None, time_limit=0.5):
        self.name = name
        self.time_limit = time_limit
        self.eval_cache = {}
        self.killer_moves = {}
        self.nodes_searched = 0

        if genome is None:
            self.genome = {
                'pawn': random.uniform(80, 120),
                'knight': random.uniform(280, 360),
                'bishop': random.uniform(290, 370),
                'rook': random.uniform(450, 550),
                'queen': random.uniform(800, 1000),
                'mobility': random.uniform(0.5, 3.0),
                'knight_center': random.uniform(10, 40),
                'pawn_center': random.uniform(10, 30),
                'bishop_pair': random.uniform(30, 80),
                'passed_pawn': random.uniform(20, 60),
                'king_safety': random.uniform(10, 50)
            }
        else:
            self.genome = genome.copy()

    def evaluate_board(self, board):
        hash_key = chess.polyglot.zobrist_hash(board)
        if hash_key in self.eval_cache:
            return self.eval_cache[hash_key]

        if board.is_checkmate():
            return -99999

        if board.is_stalemate() or board.is_insufficient_material():
            return 0

        score = 0
        center_squares = {chess.D4, chess.E4, chess.D5, chess.E5}

        piece_values = {
            chess.PAWN: self.genome['pawn'],
            chess.KNIGHT: self.genome['knight'],
            chess.BISHOP: self.genome['bishop'],
            chess.ROOK: self.genome['rook'],
            chess.QUEEN: self.genome['queen'],
            chess.KING: 20000
        }

        white_bishops = black_bishops = 0

        for square in chess.SQUARES:
            piece = board.piece_at(square)
            if piece:
                value = piece_values[piece.piece_type]

                # Позиційні бонуси
                if piece.piece_type == chess.KNIGHT and square in center_squares:
                    value += self.genome['knight_center']
                elif piece.piece_type == chess.PAWN and square in center_squares:
                    value += self.genome['pawn_center']
                elif piece.piece_type == chess.BISHOP:
                    if piece.color == chess.WHITE: white_bishops += 1
                    else: black_bishops += 1

                if piece.color == board.turn: score += value
                else: score -= value

        if white_bishops >= 2: score += self.genome['bishop_pair'] if board.turn == chess.WHITE else -self.genome['bishop_pair']
        if black_bishops >= 2: score += self.genome['bishop_pair'] if board.turn == chess.BLACK else -self.genome['bishop_pair']

        score += board.legal_moves.count() * self.genome['mobility']

        self.eval_cache[hash_key] = score
        return score

    def order_moves(self, board, depth, only_captures=False):
        moves = list(board.generate_legal_captures()) if only_captures else list(board.legal_moves)

        killers = self.killer_moves.get(depth, [])

        def move_score(move):
            score = 0
            if board.is_capture(move):
                victim = board.piece_at(move.to_square)
                attacker = board.piece_at(move.from_square)
                if victim and attacker:
                    score += 100 * victim.piece_type - attacker.piece_type
            elif move in killers:
                score += 50

            if move.promotion: score += 900
            return score

        return sorted(moves, key=move_score, reverse=True)

    def quiescence_search(self, board, alpha, beta):
        self.nodes_searched += 1
        stand_pat = self.evaluate_board(board)

        if stand_pat >= beta: return beta
        if alpha < stand_pat: alpha = stand_pat

        for move in self.order_moves(board, depth=0, only_captures=True):
            board.push(move)
            score = -self.quiescence_search(board, -beta, -alpha)
            board.pop()

            if score >= beta: return beta
            if score > alpha: alpha = score

        return alpha

    def negamax(self, board, depth, ply_from_root, alpha, beta, end_time):
        if time.time() > end_time:
            raise TimeoutError()

        self.nodes_searched += 1

        if depth == 0 or board.is_game_over():
            return self.quiescence_search(board, alpha, beta)

        ordered_moves = self.order_moves(board, depth)
        if not ordered_moves:
            return self.evaluate_board(board)

        max_val = float('-inf')

        for move in ordered_moves:
            board.push(move)
            try:
                score = -self.negamax(board, depth - 1, ply_from_root + 1, -beta, -alpha, end_time)
            finally:
                board.pop()

            if score > max_val: max_val = score
            if score > alpha: alpha = score

            if alpha >= beta:
                if not board.is_capture(move):
                    if depth not in self.killer_moves: self.killer_moves[depth] = []
                    if move not in self.killer_moves[depth]:
                        self.killer_moves[depth].insert(0, move)
                        self.killer_moves[depth] = self.killer_moves[depth][:2]
                break

        return max_val

    def choose_move(self, board):
        self.eval_cache.clear()
        self.killer_moves.clear()
        self.nodes_searched = 0

        start_time = time.time()
        end_time = start_time + self.time_limit

        best_move = None
        best_move_overall = None

        legal_moves = self.order_moves(board, depth=0)
        if not legal_moves: return None

        if len(legal_moves) == 1: return legal_moves[0]

        try:
            for current_depth in range(1, 10):
                best_value = float('-inf')
                alpha = float('-inf')
                beta = float('inf')

                if best_move_overall:
                    if best_move_overall in legal_moves:
                        legal_moves.remove(best_move_overall)
                        legal_moves.insert(0, best_move_overall)

                for move in legal_moves:
                    board.push(move)
                    try:
                        board_value = -self.negamax(board, current_depth - 1, 1, -beta, -alpha, end_time)
                    finally:
                        board.pop()

                    if board_value > best_value:
                        best_value = board_value
                        best_move = move

                    alpha = max(alpha, best_value)

                best_move_overall = best_move

        except TimeoutError:
            pass

        return best_move_overall if best_move_overall else random.choice(legal_moves)

    @staticmethod
    def crossover(parent1, parent2):
        child_genome = {}
        for key in parent1.genome.keys():
            child_genome[key] = parent1.genome[key] if random.random() < 0.5 else parent2.genome[key]
        return child_genome

    @staticmethod
    def mutate(genome, mutation_rate=0.15, mutation_strength=0.15):
        mutated = genome.copy()
        for key in mutated.keys():
            if random.random() < mutation_rate:
                mutated[key] *= random.uniform(1 - mutation_strength, 1 + mutation_strength)
                mutated[key] = max(0, mutated[key])
        return mutated

def play_match(args):
    bot1_genome, bot2_genome, game_num = args
    bot1 = GeneticBot("White", genome=bot1_genome, time_limit=0.1)
    bot2 = GeneticBot("Black", genome=bot2_genome, time_limit=0.1)

    board = chess.Board()
    move_count = 0

    while not board.is_game_over(claim_draw=True) and move_count < 80:
        bot = bot1 if board.turn == chess.WHITE else bot2
        move = bot.choose_move(board)
        if move:
            board.push(move)
            move_count += 1
        else:
            break

    if board.is_checkmate():
        return 1.0 if not board.turn else 0.0
    return 0.5

def tournament_selection(population, fitness_scores, tournament_size=3):
    selected_indices = random.sample(range(len(population)), tournament_size)
    best_idx = max(selected_indices, key=lambda idx: fitness_scores[idx])
    return population[best_idx]

def evolve_population(population_size=10, generations=5, games_per_matchup=2):
    print(f"Початок паралельної еволюції: {population_size} ботів, {generations} поколінь")
    print(f"Використовується ядер CPU: {multiprocessing.cpu_count()}\n")

    population = [GeneticBot(f"Bot_{i}") for i in range(population_size)]

    for gen in range(generations):
        print(f"{'='*40}\nПокоління {gen + 1}/{generations}\n{'='*40}")
        fitness = [0.0] * population_size

        matchups = []
        matchup_indices = []

        for i in range(population_size):
            for j in range(i + 1, population_size):
                for game_num in range(games_per_matchup):
                    if game_num % 2 == 0:
                        matchups.append((population[i].genome, population[j].genome, game_num))
                        matchup_indices.append((i, j)) # i грає білими
                    else:
                        matchups.append((population[j].genome, population[i].genome, game_num))
                        matchup_indices.append((j, i)) # j грає білими

        with ProcessPoolExecutor() as executor:
            results = list(executor.map(play_match, matchups))

        for idx, result in enumerate(results):
            white_idx, black_idx = matchup_indices[idx]
            if result == 1.0: fitness[white_idx] += 1.0
            elif result == 0.0: fitness[black_idx] += 1.0
            else:
                fitness[white_idx] += 0.5
                fitness[black_idx] += 0.5

        ranked = sorted(enumerate(fitness), key=lambda x: x[1], reverse=True)
        print("Топ-3 боти:")
        for idx, (bot_idx, score) in enumerate(ranked[:3]):
            print(f"  {idx+1}. {population[bot_idx].name}: {score:.1f} балів")

        if gen < generations - 1:
            new_population = [population[ranked[0][0]], population[ranked[1][0]]] # Елітизм
            while len(new_population) < population_size:
                parent1 = tournament_selection(population, fitness)
                parent2 = tournament_selection(population, fitness)
                child_genome = GeneticBot.mutate(GeneticBot.crossover(parent1, parent2))
                new_population.append(GeneticBot(f"Bot_Gen{gen+2}_{len(new_population)}", child_genome))
            population = new_population

    best_idx = max(enumerate(fitness), key=lambda x: x[1])[0]
    best_bot = population[best_idx]
    print("\nЕволюція завершена! Найкращий геном:")
    for k, v in best_bot.genome.items(): print(f"  {k}: {v:.1f}")
    return best_bot

def save_genome(genome, filename="champion_genome.json"):
    """Зберігає геном бота у файл."""
    with open(filename, 'w', encoding='utf-8') as f:
        # indent=4 робить файл красивим і читабельним для людини
        json.dump(genome, f, indent=4)
    print(f"\n[ЗБЕРЕЖЕННЯ] Геном успішно збережено у файл '{filename}'")

def load_genome(filename="champion_genome.json"):
    """Завантажує геном з файлу, якщо він існує."""
    if os.path.exists(filename):
        with open(filename, 'r', encoding='utf-8') as f:
            genome = json.load(f)
        print(f"\n[ЗАВАНТАЖЕННЯ] Геном успішно прочитано з '{filename}'")
        return genome
    else:
        print(f"\n[ІНФО] Файл '{filename}' не знайдено. Буде використано випадковий геном.")
        return None
```
